```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('baseball.csv')
df.head()
```

```python
df['RD'] = df['RS']- df['RA']
df.head()
```

```python
moneyball0 = df[df.Playoffs == 0]
moneyball0.head()
```

```python
moneyball1 = df[df.Playoffs == 1]
moneyball1.head()
```

```python
#wins vs runs scored for playoffs teams

plt.figure(figsize=(10,8))
plt.scatter(x=moneyball0.W, y=moneyball0.RS, c='red')
plt.scatter(x=moneyball1.W, y=moneyball1.RS, c='blue')
plt.xlabel("Wins")
plt.ylabel("Runs Scored")
plt.axvline(x = 99)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x=moneyball0.W, y=moneyball0.OBP, c='red')
plt.scatter(x=moneyball1.W, y=moneyball1.OBP, c='blue')
plt.xlabel("Wins")
plt.ylabel("OBP")
plt.axvline(x = 99)
plt.axhline(y = .343)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x=moneyball0.W, y=moneyball0.RD, c='red')
plt.scatter(x=moneyball1.W, y=moneyball1.RD, c='blue')
plt.xlabel("Wins")
plt.ylabel("RD")
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x=moneyball0.W, y=moneyball0.OOBP, c='red')
plt.scatter(x=moneyball1.W, y=moneyball1.OOBP, c='blue')
plt.xlabel("Wins")
plt.ylabel("RD")
```
